<h2>Python Function Implementing an Algorithm To Determine if Two Graphs are Isomorphic or Not</h2>

```python
# Plots.
import matplotlib.pyplot as plt
# Resize plots.
plt.rcParams['figure.figsize'] = (6, 6)
# Matrices.
import numpy as np
# Permutations
import itertools as it
# Graphs.
import networkx as nx
# Math.
import math
# Networkx drawing parameters.
params = {'node_color': 'lightskyblue', 'node_size': 800, 'font_size': 18, 'with_labels': True}
```

```python
# Build 1st Graph
G1=nx.Graph()
G1.nodes(['KA','TN','TL','AP','KL','GO'])
G1.add_edges_from([('KA','TN'),('KA','TL'),('KA','TL'),('KA','AP'),('KA','KL'),('KA','GO')])
```

```python
#Draw the 1st graph
nx.draw(G1,with_labels=1)
```

```python
#Create Graph 2
G2=nx.star_graph(5)
#Draw Graph 2
nx.draw(G2, with_labels=1)
```

```python
nx.is_isomorphic(G1,G2)
```

```python
##The 2 Graphs are isomorphic
```
